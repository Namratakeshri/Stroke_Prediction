4. Define a new column (in original df) that lables the records based on the 'avg_glucose_level'. Name this new column 'classification'. Use the following: <br>
if x < 125; normal<br>
if 125 =< x <= 200; high<br>
if x > 200; very high<br>
(x is 'avg_glucose_level')

```python
# create a list of our conditions
conditions = [
    (stroke_df['avg_glucose_level'] <= 125),
    (stroke_df['avg_glucose_level'] > 125) & (stroke_df['avg_glucose_level'] <= 200),
    (stroke_df['avg_glucose_level'] > 200) ,
     ]

# create a list of the values we want to assign for each condition
values = ['normal', 'high', 'very high']

# create a new column and use np.select to assign values to it using our lists as arguments
stroke_df['classification'] = np.select(conditions, values)
```
